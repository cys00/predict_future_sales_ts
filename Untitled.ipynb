```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
train = pd.read_csv('sales_train.csv',parse_dates = ['date'], infer_datetime_format = True, dayfirst = True)
#train['date'] = train['date'].apply(lambda s: datetime.strptime(s,'%d.%m.%Y')) 
item_categories = pd.read_csv('item_categories.csv')
items = pd.read_csv('items.csv')
sample_submission = pd.read_csv('sample_submission.csv')
shops = pd.read_csv('shops.csv')
test = pd.read_csv('test.csv')
```

```python
train = train[train['item_cnt_day']>0]
```

```python
train.head()
```

```python
train.dtypes
```

```python
data = train.groupby(['shop_id','item_id','date_block_num']).sum().reset_index()
data.head()
```

```python
data = data[['shop_id','item_id','date_block_num','item_cnt_day']]
```

```python
data.head()
```

```python
#transform to time series data 
data = pd.pivot_table(data,index=['shop_id','item_id'],columns='date_block_num',values='item_cnt_day',fill_value=0).reset_index()
```

```python
data.head()
```

```python
data.shape
```

```python
data = pd.merge(test,data,on=['shop_id','item_id'],how='left')
```

```python
data.fillna(0,inplace=True)
```

```python
data.shape
```

```python
data.head()
```

```python
x_train = data.drop(columns=['shop_id','item_id',33])
y_train = data[33]
```

```python
x_train.shape
```

```python
x_test = data.drop(columns=['shop_id','item_id',0])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size = 0.2, random_state = 0)
```

```python
from lightgbm import LGBMRegressor
```

```python
model_lgb = LGBMRegressor( n_estimators=200,
                           learning_rate=0.03,
                           num_leaves=32,
                           colsample_bytree=0.9497036,
                           subsample=0.8715623,
                           max_depth=8,
                           reg_alpha=0.04,
                           reg_lambda=0.073,
                           min_split_gain=0.0222415,
                           min_child_weight=40)
model_lgb.fit(x_train, y_train)
y_pred_lgb = model_lgb.predict(x_test)
```
